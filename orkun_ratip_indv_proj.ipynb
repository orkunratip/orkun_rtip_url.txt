```python
partArray=[]
listofTimeOfDay=timestamp_date.dt.hour
for i in range (len(df["pickup_datetime"])):
    if 7<listofTimeOfDay[i]<=9:  
        part="rush_hour_morning"
    elif 9<listofTimeOfDay[i]<=16:
        part="afternoon"
    elif 16<listofTimeOfDay[i]<=18: 
        part="rush_hour_evening"
    elif 18<listofTimeOfDay[i]<=23:
        part="evening"
    elif 23<listofTimeOfDay[i]<=24: 
        part="late_night"    
    elif 0<listofTimeOfDay[i]<=7: 
        part="late_night"   
    partArray.append(part)
df["time_of_day"]=partArray
df["time_of_day"]
```

```python
test_type = "distance"
singlePassenger = df[df["passenger_count"] ==1][test_type]  
multiplePassenger = df[df["passenger_count"]>1][test_type]  
meanofsinglepassenger=singlePassenger.mean()
meanofmultiplepassenger=multiplePassenger.mean()
objects = ('single', 'multiple')
y_pos = np.arange(len(objects))
means=[meanofsinglepassenger,meanofmultiplepassenger]
plt.barh(y_pos, means, align='center', alpha=0.5)
plt.yticks(y_pos, objects)
plt.xlabel('distance')
plt.title('Passenger count and distance means correlation')
plt.show()
```

grouping the number of passengers into 2 as single passengers and multiple passengers. taking the means of each groups distances and plotting the distance and group size correlation by using bar chart. 

```python
stats.ttest_ind(singlePassenger, multiplePassenger, equal_var=False) 
```

Since our p-value is smaller than 0.05, we can argue that there is significant difference between the distance mean of single passenger and the distance mean of multiple passengers. Hence we would reject our Null Hypothesis meaning that there is a positive correlation between distance and number of passengers.

```python
dayArray=[]
timestamp_day= pd.to_datetime(df["pickup_datetime"], infer_datetime_format=True)
day=timestamp_day.dt.day_name()
for i in range (len(df["pickup_datetime"])):
    if day[i]=="Sunday":
        theDay="weekend"
    elif day[i]=="Saturday":
        theDay="weekend"
    else:
        theDay="weekday"
        
    dayArray.append(theDay)
df['weekDays']=dayArray
df['weekDays']
```

grouping the days in a week into two as weekend and weekdays and creating a new column for the information.

```python
print(df.columns)
```

showing the new added column weekDays

```python
test_type = "distance"
WEEKDAYS = df[df["weekDays"] =="weekday"][test_type]  
WEEKEND = df[df["weekDays"]=="weekend"][test_type]
meanofweekday=WEEKDAYS.mean()
meanofweekend=WEEKEND.mean()
objects = ('weekday', 'weekend')
y_pos = np.arange(len(objects))
means=[meanofweekday,meanofweekend]
plt.barh(y_pos, means, align='center', alpha=0.5)
plt.yticks(y_pos, objects)
plt.xlabel('week')
plt.title('weekend/weekday and distance means correlation')
plt.show()
```

grouping the week into 2 as weekend and weekdays. taking the means of each groups distances and plotting the distance and part of week correlation by using bar chart. 

```python
stats.ttest_ind(WEEKDAYS, WEEKEND, equal_var=False) 
```

Since our p-value is smaller than 0.05, we can argue that there is significant difference between the distance mean of weekends and the distance mean of weekdays. Hence we would reject our Null Hypothesis meaning that there is a positive correlation between distance and days of week.
